Q6. Median horsepower before/after fill

```python
# Original median
print("Before:", df['horsepower'].median())

# Most frequent value (mode)
most_frequent = df['horsepower'].mode()[0]

# Fill NaN with mode
df['horsepower'] = df['horsepower'].fillna(most_frequent)

# New median
print("After:", df['horsepower'].median())
```

Q7. Sum of weights (mini regression)

```python
import numpy as np

# Select Asia cars
df_asia = df[df['origin'] == 'Asia'][['vehicle_weight', 'model_year']]

# First 7 rows
X = df_asia.head(7).values

# XTX = X^T * X
XTX = X.T.dot(X)

# Invert XTX
XTX_inv = np.linalg.inv(XTX)

# y vector
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])

# w = (XTX^-1 * X^T * y)
w = XTX_inv.dot(X.T).dot(y)

print(w.sum())
```
